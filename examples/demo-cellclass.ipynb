# Demo of using cell class
This demo will demonstrate how to use the `cellclass`.

Load cell data into `Cell` class using `loaddatatools` module.

```python
# import `loaddatatools` to load data into cell class
import loaddatatools as ldt
from matplotlib import pyplot as plt
import socket
import numpy as np
%matplotlib inline
```

#####Load all cells

```python
# Load all cells to the `cells` dictionary
cells = ldt.loadcells()
```

```python
print(cells.keys())
```

#####Load one cell

```python
cell = ldt.loadcell('g9')
cell?
```

##### Plots

```python
sec_length = 20000
contrast = np.ones(len(cell.contrast) * sec_length)
i = 0
for c in cell.contrast:
    contrast[i*sec_length:(i+1)*sec_length] = c * np.ones(sec_length)
    i += 1

fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(411)
ax.plot(contrast, 'k')
ax.set_xlabel('$Time$', fontsize=20)
ax.set_ylabel('$\sigma/\mu$', fontsize=20)
ax.set_title('Contrast', fontsize=20, y=1.05)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)

ax = fig.add_subplot(412)
ax.plot(cell.st, 'k')
ax.set_xlabel('$Time$', fontsize=20)
ax.set_ylabel('$Intensity$', fontsize=20)
ax.set_title('Stimulus', fontsize=20, y=1.05)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)

ax = fig.add_subplot(413)
ax.plot(cell.mp - np.mean(cell.mp), 'b')
ax.set_xlabel('$Time$', fontsize=20)
ax.set_ylabel('$mV$', fontsize=20)
ax.set_title('Membrane Potential', fontsize=20, y=1.05)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)

ax = fig.add_subplot(413)
ax.plot(cell.mp, 'b')
ax.set_xlabel('$Time$', fontsize=20)
ax.set_ylabel('$mV$', fontsize=20)
ax.set_title('Membrane Potential', fontsize=20, y=1.05)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)

ax = fig.add_subplot(414)
ax.plot(cell.fr, 'r')
ax.set_xlabel('$Time$', fontsize=20)
ax.set_ylabel('$Hz$', fontsize=20)
ax.set_title('Firing Rate', fontsize=20, y=1.05)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)

fig.tight_layout()
plt.show()
```
